```python
import pandas as pd

emea1 = pd.read_csv('Master 1 EMEA.csv', delimiter=',', encoding='ISO-8859-1')
emea1.head()
emea2 = pd.read_csv('Master 2 EMEA.csv', delimiter=',', encoding='ISO-8859-1')
emea = pd.concat([emea1, emea2], ignore_index=True)
emea1.info()
emea2.info()
emea = emea.drop(emea.columns[39:45], axis = 1)
emea.info()
emea.isnull().sum()

#Clean date
emea['Day'].unique()
emea['clean_day'] = emea['Day'].astype(str).apply(lambda x: x.split(' ')[0] if ' ' in x else x)
emea['clean_day'] = emea['clean_day'].astype(str).apply(lambda x: str(x).zfill(2))
emea['clean_month'] = emea['Fiscal month'].astype(str).apply(lambda x: x.split(' ')[0] if ' ' in x else x)
emea['clean_month'].unique()
emea['clean_month'] = emea['clean_month'].astype(str).apply(lambda x: str(x).zfill(2))
emea['cleaned_date'] = '2022-'+emea['clean_month']+'-'+emea['clean_day']
emea['cleaned_date'].unique()
emea['cleaned_date'] = pd.to_datetime(emea['cleaned_date'], format = '%Y-%m-%d', errors='coerce')
print(emea['cleaned_date'].dtype)
emea['cleaned_month'] = emea['cleaned_date'].apply(lambda x: x.month)
emea['cleaned_day'] = emea['cleaned_date'].apply(lambda x: x.day)
emea['cleaned_date'].isnull().sum()
##46,087 is null

#Drop unneccesary columns
emea.drop(['DODTYP', 'REPEAT Acc. No.'], axis = 1)
emea.info()

#LIST OF COLUMNS NAME TO CALL FROM
print(list(emea.columns))

#Convert data type
##Remove any strange characters that is not a digit, decimal points; then connect the rest together
emea[' VALUE \x80 '].unique()
def clean_string(s):
    if isinstance(s, str):
        # Remove any character that is not a digit, decimal point, or negative sign
        cleaned = ''.join(char for char in s if char.isdigit() or char == '.') #or char == '-'
        return cleaned
    else:
        return s
emea['value_convert'] = emea[' VALUE \x80 '].apply(clean_string)
#Check and count the number of empty-string cells (empty cells but was not count as NaN)
empty_string_count = emea['value_convert'].apply(lambda x: x == '').sum()

##Convert to numeric and raise the error where there are unconvertible values
def safe_to_numeric(data):
    try:
        return pd.to_numeric(data, errors='raise')
    except ValueError as e:
        print(f"ValueError occurred: {e}")
        # Handle the error, for example by coercing invalid entries to NaN
        return pd.to_numeric(data, errors='raise')
emea['value_float'] = emea['value_convert'].apply(safe_to_numeric)
emea['value_float'].dtype

#Convert product volume per piece column
emea['volume_convert'] = emea[' Product volume (m3 / pcs) '].apply(clean_string)
empty_string_count_vol = emea['volume_convert'].apply(lambda x: x == '').sum()
print(empty_string_count_vol)

###END OF THE CLEANING. Next step we just chose the columns that are relevant for the analysis to work with.

#Select relevant columns to work with
order_pattern = emea[['Account Number',
                      'Order Number','SKU Number', 'Quantity', 'Sales Unit of Measure', 'Sales Order Origin',
                      ' VALUE \x80 ',' Product weight (kg. / pcs) ',' Product volume (m3 / pcs) ',
                      'Product category', 'Product Segment', 'Destination country', 'Customer type',
                      'cleaned_date', 'cleaned_month', 'cleaned_day']]
order_pattern.info()
order_pattern.isnull().sum()
sales_pattern = order_pattern[['Order Number','SKU Number', 'Quantity', 'Sales Unit of Measure',
                      ' VALUE \x80 ',' Product weight (kg. / pcs) ',' Product volume (m3 / pcs) ',
                      'Product category', 'Product Segment', 'Customer type',
                      'cleaned_date', 'cleaned_month', 'cleaned_day']]
sales_pattern.info()
sales_pattern.isnull().sum()
sales_pattern.dropna()


#Order frequency per customer types
channel_count = sales_pattern.pivot_table(
    index='Customer type',
    values='Order Number',
    aggfunc=lambda x: x.nunique()
).reset_index()

q1 = sales_pattern[sales_pattern['cleaned_month'].isin([1, 2, 3])]
q1_order_counts = q1.groupby('Customer type')['Order Number'].nunique().reset_index()
sales_pattern['Customer type'].unique()
shop_type_3 = sales_pattern[sales_pattern['Customer type'] == "SHOP TYPE 3"]
personalized_material = sales_pattern[sales_pattern['Customer type'] == "PERSONALIZED MATERIAL"]
dist_exp = sales_pattern[sales_pattern['Customer type'] == "DISTRIBUTOR / EXPORT"]
shop_type_2 = sales_pattern[sales_pattern['Customer type'] == "SHOP TYPE 2"]
shop_type_1 = sales_pattern[sales_pattern['Customer type'] == "SHOP TYPE 1"]
mail_order = sales_pattern[sales_pattern['Customer type'] == "MAIL ORDER"]
intercompany = sales_pattern[sales_pattern['Customer type'] == "INTERCOMPANY"]
department_store = sales_pattern[sales_pattern['Customer type'] == "DEPARTMENT STORE"]
track = sales_pattern[sales_pattern['Customer type'] == "TRACK"]
cross_market = sales_pattern[sales_pattern['Customer type'] == "CROSS MARKET CUSTOMER"]
promotion = sales_pattern[sales_pattern['Customer type'] == "PROMOTION"]
misc = sales_pattern[sales_pattern['Customer type'] == "MISC"]
internet_acc = sales_pattern[sales_pattern['Customer type'] == "INTERNET ACCOUNT"]
sport_allowance = sales_pattern[sales_pattern['Customer type'] == "SPORT ALLOWANCE"]
own_use = sales_pattern[sales_pattern['Customer type'] == "OWN USE"]
direct_cus = sales_pattern[sales_pattern['Customer type'] == "DIRECT TO CONSUMER"]
```
